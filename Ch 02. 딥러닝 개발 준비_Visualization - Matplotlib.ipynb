#### 막대그래프 bar plot : ax.bar
bar plot은 ax.bar와 같이 사용하며 다음과 같은 인자를 받습니다
- x: x축에 사용될 값
- height: y축에 사용될 값
- width: 각 bar의 width
- bottom: y축의 base 값
- align: 정렬(center or edge)

```python
# bar plot : 막대그래프
fig = plt.figure()
ax = fig.add_subplot()
ax.bar([0, 1, 2, 3, 4], [3, 2, 5, 4, 6])
plt.show()

#필사--------------------------------------------------------------------------------
#bar plot : 막대 그래프
fig = plt.figure()
ax = fig.add_subplot()
ax.bar([0, 1, 2, 3, 4], [3, 2, 5, 4, 6])
# ax.bar([0, 1, 2, 3, 4], [3, 2, 5, 4, 6], width=0.5)
# ax.bar([0, 1, 2, 3, 4], [3, 2, 5, 4, 6], bottom=2)
# ax.bar([0, 1, 2, 3, 4], [3, 2, 5, 4, 6], width=0.5)
# ax.bar([0, 1, 2, 3, 4], [3, 2, 5, 4, 6], align='edge')
plt.show()
```

```python
# width 조정
fig = plt.figure()
ax = fig.add_subplot()
ax.bar([0, 1, 2, 3, 4], [3, 2, 5, 4, 6], width=0.5)
plt.show()
```

```python
# bottom 조정 : y축 2부터 시작
fig = plt.figure()
ax = fig.add_subplot()
ax.bar([0, 1, 2, 3, 4], [3, 2, 5, 4, 6], bottom=2)
plt.show()
```

```python
# align 조정
fig = plt.figure()
ax = fig.add_subplot()
ax.bar([0, 1, 2, 3, 4], [3, 2, 5, 4, 6], align='edge') #엣지정렬 = 가장자리정렬
plt.show()
```

```python
ax.bar() #노트북은 되는데 내 pc에선 안됨ㅜ
#ax.bar(shift+tab)
'''
def bar(x, height, width=0.8, bottom=None, *, align='center', **kwargs)
Make a bar plot.

The bars are positioned at *x* with the given *align*\ment. Their
dimensions are given by *width* and *height*. The vertical baseline
is *bottom* (default 0).

Each of *x*, *height*, *width*, and *bottom* may either be a scalar
applying to all bars, or it may be a sequence of length N providing a
separate value for each bar.

Parameters

x : sequence of scalars
    The x coordinates of the bars. See also *align* for the
alignment of the bars to the coordinates.

height : scalar or sequence of scalars
    The height(s) of the bars.

width : scalar or array-like, optional
    The width(s) of the bars (default: 0.8).

bottom : scalar or array-like, optional
    The y coordinate(s) of the bars bases (default: 0).

align : {'center', 'edge'}, optional, default: 'center'
    Alignment of the bars to the *x* coordinates:


'center': Center the base on the *x* positions.

'edge': Align the left edges of the bars with the *x* positions.


    To align the bars on the right edge pass a negative *width* and
align='edge'.
'''
```

#### 히스토그램 histogram plot : ax.hist

histogram은 data의 분포를 나타낼 때 사용하며, 주로 다음과 같은 인자를 사용합니다.
- x: 입력 data
- bins: histogram을 나타낼 x축의 구간, 붙어있는 막대의 갯수, 숫자가 높아질수록 촘촘해짐
  (정수 입력시 구간의 갯수, list와 같은 sequence 입력시는 sequence 값이 구간으로 설정됨)
- range: x축 범위 (x.min, x.max)로 표현
- density: 확률 분포로 나타낼 것인지 여부

```python
# histogram
np.random.seed(1111) 

fig = plt.figure()
ax = fig.add_subplot()
ax.hist(np.random.randn(500))
plt.show()

#필사--------------------------------------------------------------------------------

np.random.seed(1111) #np.random.seed(42) => 난수생성코드, 안의 숫자가 0,42,1111처럼 달라지면 정해진 알고리즘에 따라 난수도 바뀜

fig = plt.figure()
ax = fig.add_subplot()
ax.hist(np.random.randn(500)) # 평균0, 표준정규분포1인 (가우시안)표준정규분포 난수 500개 생성, (m, n)명시에는 m행 n열 난수생성
#※ np.random.rand(6) 0~1 균일분포 난수생성
#※ np.random.randint(6) 0~5까지의 랜덤정수 1개가 뽑힘, (m, n) m~n-1사이의 숫자 1개 뽑힘 
plt.show()
```

```python
# bins : histogram을 나타낼 x축의 구간, 붙어있는 막대의 수
np.random.seed(1111)

fig = plt.figure()
ax = fig.add_subplot()
ax.hist(np.random.randn(500), bins=50) #bins=50 : 붙어있는 막대가 50개임
plt.show()
```

```python
# bins 
np.random.seed(1111)

fig = plt.figure()
ax = fig.add_subplot()
#ax.hist(np.random.randn(500), [-3,-2,-1,0,1,2,3]) #  [-3,-2,-1,0,1,2,3]: x축의 범위
#ax.hist(np.random.randn(500), [-2,-1,0,1,2,3])
ax.hist(np.random.randn(500), np.linspace(-2, 2, num=40)) #np.linspace(-2, 2, num=40) : x축을 -2~2까지 40구간을 나눠서 표현해라.(붙어있는 막대가 40개임)
plt.show()
```

```python
# range 
np.random.seed(1111)

fig = plt.figure()
ax = fig.add_subplot()
ax.hist(np.random.randn(500), bins=40, range=(-2,2))
plt.show()
```

```python
# density: 확률분포가 필요한 경우
np.random.seed(1111)

fig = plt.figure()
ax = fig.add_subplot()
ax.hist(np.random.randn(500), bins=40, range=(-2,2), density=True) #안에 면적이 1인 확률이 되도록 y축 값이 변함, 확률분포가 필요한 경우 True로 사용
plt.show()
```

```python
import os
os.getcwd()
```

```python
cur_dir = os.getcwd() #/content
data_dir = os.path.join(cur_dir, 'oxford_pet') 
image_dir = os.path.join(data_dir, 'images') #옥스포트 펫 폴더 아래 images 폴더 이미지 경로가 있음
bbox_dir = os.path.join(data_dir, 'annotations', 'xmls') #바운딩박스 경로(바운딩박스 정보)는 옥스포트 펫 폴더 아래 annotations 폴더 아래 xmls 폴더에 있음
seg_dir = os.path.join(data_dir, 'annotations', 'trimaps') #세그멘테이션 경로(세그멘테이션 정보)는 옥스포트 펫 폴더 아래 annotations 폴더 아래 trimaps 폴더에 있음
```

```python
# 파일을 리스트 형태로 만들기
image_files = [fname for fname in os.listdir(image_dir) if os.path.splitext(fname)[-1] == '.jpg'] 
bbox_files = [fname for fname in os.listdir(bbox_dir) if os.path.splitext(fname)[-1] == '.xml']
seg_files = [fname for fname in os.listdir(seg_dir) if os.path.splitext(fname)[-1] == '.png']
```

```python
rnd_idx = random.randint(1, len(bbox_files)) #bbox_files 바운딩 박스 갯수 중 1개를 랜덤하게 뽑음, 바운딩 박스가 갯수가 제일 적어서 예시로 들었음
bbox_name = bbox_files[rnd_idx] # 랜덤하게 뽑은 바운딩 박스의 이름을 가져옴,                               
bbox_path = os.path.join(bbox_dir, bbox_name) # 예를 들어 Abyssinian_10.xml이 뽑혔다고 하면 bbox_dir에 다가 Abyssinian_10.xml을 조인을 하면 

#print(bbox_path) # 코드실행 할때마다 값이 바뀜
# /content/oxford_pet/annotations/xmls/pomeranian_126.xml
# /content/oxford_pet/annotations/xmls/newfoundland_137.xml
```

```python
# xml 파싱하는 법 
# import xml.etree.ElementTree as et
tree = et.parse(bbox_path)

width = float(tree.find('./size/width').text)
height = float(tree.find('./size/height').text)
xmin = float(tree.find('./object/bndbox/xmin').text) #xml파일을 notepad로 열어보면 <object>아래 <bndbox>가 나오고 그아래 <xmin></xmin> 형태로 있음
                                                     #텍스트로 들고와서 float로 바꿔 들고와서
xmax = float(tree.find('./object/bndbox/xmax').text) #xml파일을 notepad로 열어보면 <object>아래 <bndbox>가 나오고 그아래 <xmax></xmax> 형태로 있음
ymin = float(tree.find('./object/bndbox/ymin').text) #xml파일을 notepad로 열어보면 <object>아래 <bndbox>가 나오고 그아래 <ymin></ymin> 형태로 있음
ymax = float(tree.find('./object/bndbox/ymax').text) #xml파일을 notepad로 열어보면 <object>아래 <bndbox>가 나오고 그아래 <ymax></ymax> 형태로 있음

rect_x = xmin #
rect_y = ymin
rect_w = xmax-xmin #가로길이
rect_h = ymax-ymin #세로길이

fname = os.path.splitext(bbox_name)[0]+'.jpg' #ex> Abyssinian_10.xml를 Abyssinian_10.jpg 형태로 확장자 바꾸기, 여기서 [0]은 파일명(Abyssinian_10)을 의미함
                                              #이렇게 바꾸면 xml이 jpg이미지가 됨
fpath = os.path.join(image_dir, fname)
image = Image.open(fpath) #이미지를 열고 
# image = np.array(image)

bbox_name = os.path.splitext(fname)[0]+'.xml'

fig, ax = plt.subplots(1, 1)
#사각형을 Rectangle이라는 라이브러리를 통해서 그릴 것임
rect = Rectangle((rect_x, rect_y), rect_w, rect_h, fill=False, color='red')
                # (왼쪽 꼭지점 좌표), 가로길이, 세로길이, 사각형 안을 채울건지(박스를 그릴 것이기에 채우면안됨), 
ax.add_patch(rect) #패치안에 렉탱글을 집어넣고 
ax.imshow(image) #imshow를 하면
plt.show()
```

```python
sname = os.path.splitext(bbox_name)[0]+'.png' #png로 바꾸고
spath = os.path.join(seg_dir, sname)
seg = Image.open(spath)

fig, ax = plt.subplots(1, 2, figsize=(15,8)) #1행 2열: 옆으로 2개 이미지를 그림
ax[0].imshow(image) #왼쪽은 이미지
ax[1].imshow(seg)   #오른쪽은 세그멘테이션 맵을 그림
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.imshow(image) #아임쇼를 두번하면 겹쳐나옴, 덮어씀, 따라서 알파값을 줌
ax.imshow(seg, alpha=0.5) #알파값은 투명도, 50% 투명하게 보여줘
plt.show()
```
